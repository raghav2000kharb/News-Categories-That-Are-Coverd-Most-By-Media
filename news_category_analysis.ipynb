### Reading the data frame into the jupyter note book, and importing required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
##iris_data = pd.read_json("News_Category_Dataset_v2.json", lines = True, encoding="utf8")
df= pd.read_json(open("/home/raghav/Desktop/News_Category_Dataset_v2.json", "r", encoding="utf8"),lines=True)
df

```

```python
df.info()
```

### Creating a different dataframe for the combined categories

```python
cats = df.groupby(['category']).count().reset_index()
cats 

```

###  Graph for showing distribution of number of hradlines across varies news categories

```python
plt.figure(figsize=(25,20))
plt.xlabel("Number of headlines in each News Category")
plt.ylabel("Categories")
plt.barh(cats["category"],cats["headline"], align="center",color="red") 
```

###  Extracting Year from the Date

```python
df['year'] = df['date'].dt.year
df

```

### Number of Headlines in Different Categories over span of 6 Years (2012-18)

```python
years = df.groupby(['category','year']).count().reset_index()
years

```

### 2012

```python
year1= years[df['category'].notnull() & (years['year'] == 2012)]
year1
plt.figure(figsize=(25,20))
plt.xlabel("Categories")
plt.ylabel("Number of Headlines of Different Categories in 2012")
plt.bar(year1["category"],year1["headline"], align="center",color="blue") 
```
